```python
def check_falsy(name, value):
  if value:
    print(f"{name} = {value} evaluated to True")
  else:
    print(f"{name} = {value} evaluated to False")
```

```python
x = 0
y = -1
z = 3
t = 0.5
st = "djfno"
empty_str = ""
blank_str = " "

vars = {
  "x": x,
  "y": y,
  "z": z,
  "t": t,
  "st": st,
  "empty_str": empty_str,
  "blank_str": blank_str
}

for name, value in vars.items():
  check_falsy(name, value)
```
